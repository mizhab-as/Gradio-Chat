```python
import gradio as gr
import os


def respond(message, history, system_prompt, temperature):
    messages = []
    if system_prompt:
        messages.append({"role": "system", "content": system_prompt})

    for turn in history:
        # Gradio Chatbot history is a list of tuples (user_message, bot_message)
        # Convert to OpenAI-style dictionary for consistency
        messages.append({"role": "user", "content": turn[0]})
        if turn[1] is not None: # Bot response might be None during streaming
            messages.append({"role": "assistant", "content": turn[1]})

    messages.append({"role": "user", "content": message})

    from groq import Groq
    client = Groq(api_key=os.environ["GROQ_API_KEY"])

    response = client.chat.completions.create(
        model="llama-3.1-8b-instant",
        messages=messages,
        temperature=temperature,
        stream=True,
    )

    partial_message = ""
    for chunk in response:
        if chunk.choices[0].delta.content is not None:
            partial_message += chunk.choices[0].delta.content
            yield partial_message

system_prompt_textbox = gr.Textbox(
    label="System Prompt",
    value="You are a helpful AI assistant that can answer questions and engage in conversation on a wide range of topics.",
    interactive=True
)
temperature_slider = gr.Slider(
    minimum=0.0,
    maximum=2.0,
    value=0.4,
    step=0.01,
    label="Temperature (randomness)",
    interactive=True
)

# Default values for system prompt and temperature for examples
default_system_prompt = system_prompt_textbox.value
default_temperature = temperature_slider.value

demo = gr.ChatInterface(
    fn=respond,
    chatbot=gr.Chatbot(height=300, type='messages'),
    textbox=gr.Textbox(placeholder="Ask me anything!", container=False, scale=7),
    title="General Purpose Chatbot",
    description="Engage in conversation with a customizable AI assistant.",
    theme="soft",
    # Examples now include values for message, system_prompt, and temperature
    examples=[
        ["What is the capital of France?", default_system_prompt, default_temperature],
        ["Tell me a fun fact about space.", default_system_prompt, default_temperature],
        ["Explain the concept of quantum computing.", default_system_prompt, default_temperature]
    ],
    cache_examples=True,
    additional_inputs=[system_prompt_textbox, temperature_slider]
)

demo.launch(debug=True)
```
